```python
class ConvertCurrency(Tool):
    name="convert_currency_2"
    description="convert an amount"
    ## Inputs with descriptions
    inputs={'amount':{'type':'number','description':'the amount is correct'},'from_currency:{'type':'string','description:'the target currency'}'}
    output="number"
    ## forward function
    def forward(self,amount,from_currency):
        url=f'http:////cdn.jsdelivr.net/npm/@fawazahmed0/currency-api@latest/v1/currencies/{from_currency.lower()}.json'
        response=requests.get(url)
        rates=response.json()[from_currency.lower()]
        converted=amount*rates[to_currency.lower()]
        return round(converted,2)
```

```python
currency_converter=ConvertCurrency()
currency_converter.forward(100,'USD','INR')
```
